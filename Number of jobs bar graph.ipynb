```python
import pandas as pd
!pip install openpyxl
from openpyxl import Workbook
df=pd.read_excel('job-postings.xlsx')
df
```

```python
df_sorted=df.sort_values(by='Number_of_Jobs')
df_sorted.set_index('Location', inplace=True)
df_sorted
```

```python
import matplotlib.pyplot as plt
```

```python
df_sorted.plot(kind='barh', color='skyblue', alpha=.8, figsize=(12,8))
plt.title('Number of Jobs per Location')
plt.xlabel('Number of Jobs')
plt.ylabel('Location')
for i, v in enumerate(df_sorted['Number_of_Jobs']):
    plt.text(v + 0.1, i, str(v), color='black', va='center')
plt.show()
```
