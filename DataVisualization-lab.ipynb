## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt
```

```python
# your code goes here
query1 = """
select ConvertedComp from master"""

df1 = pd.read_sql_query(query1,conn)

df1.plot(kind='hist',color= 'green', figsize=(10,6), alpha=.6)
plt.title('Converted Compensation Histogram')
plt.xlabel('Comverted Compensation')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
query2 = """
select Age from master """

df2=pd.read_sql_query(query2,conn)

df2.plot(kind='box', color='green', figsize=(10,6))
plt.title('Box Plot of Age')
plt.ylabel('Years')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
query3="""
select Age, WorkWeekHrs from master 
"""
df3=pd.read_sql_query(query3, conn)

df3.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10,8), color='green', alpha=.6)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
query4="""select Age, WorkWeekHrs, CodeRevHrs from master"""
df4=pd.read_sql_query(query4, conn)

df4.plot(kind='scatter', x='WorkWeekHrs', y='CodeRevHrs', s='Age')
plt.title('Scatter Plot with Bubbles Size is Based on Age')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
query5 = """select DatabaseDesireNextYear as Technology, count(*) num_of_respondents from DatabaseDesireNextYear
group by DatabaseDesireNextYear order by  num_of_respondents desc
limit 5"""

df5=pd.read_sql_query(query5, conn)
df5
```

```python
plt.figure(figsize=(9,9))
explode = [0.1 if i == 0 else 0 for i in range(len(df5))]
plt.pie(df5['num_of_respondents'],
        labels=df5['Technology'],
        autopct='%1.1f%%',
        startangle=90,
        explode=explode,
        colors=['lightgreen', 'skyblue', 'coral', 'gold', 'orange'])
plt.axis('equal')
plt.title('Top 5 Desired DataBases to Learn Next Year')
plt.show()
```

```python
language_query = """select LanguageDesireNextYear as Language, count(*) num_of_respondents from LanguageDesireNextYear
group by LanguageDesireNextYear order by  num_of_respondents desc
limit 5"""

top_lang=pd.read_sql_query(language_query, conn)
top_lang
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
query6="""select Age, WorkWeekHrs, CodeRevHrs from master where age between 30 and 35"""
df6=pd.read_sql_query(query6, conn)
df6
median=df6.groupby('Age').median()
median

median.plot(kind='bar', stacked=True, figsize=(6,6))
plt.title('Median Work Hours by Age')
plt.xlabel('Age')
plt.ylabel('Work Hours')
plt.show()
```

### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
query7="""select Age, ConvertedComp from master where Age between 45 and 60"""
df7=pd.read_sql_query(query7, conn)
df7
df7_median=df7.groupby('Age')['ConvertedComp'].median()
df7_median

df7_median.plot(kind='line', figsize=(8,6), color='green', legend=False, marker='o', grid=True)
plt.title('Median of Converted Compensation (USD) from Ages 45 top 60')
plt.xlabel('Age')
plt.ylabel('Converted Compensation (USD)')
plt.show()
```

```python
# your code goes here
query7="""select Age, ConvertedComp from master where Age between 25 and 30"""
df71=pd.read_sql_query(query7, conn)
df71
df71_median=df71.groupby('Age')['ConvertedComp'].median()
df71_median

df71_median.plot(kind='line', figsize=(8,6), color='green', legend=False, marker='o', grid=True)
plt.title('Median of Converted Compensation (USD) from Ages 25 top 30')
plt.xlabel('Age')
plt.ylabel('Converted Compensation (USD)')
plt.show()
```

Create a horizontal bar chart using column `MainBranch.`

```python
# your code goes here
query8="""select MainBranch, count(*) as Num_of_Respondents from master group by MainBranch order by Num_of_Respondents desc """
df8=pd.read_sql_query(query8,conn)
df8
labels=['Developer by Profession', 'Not Primarily a Developer']
df8['MainBranch']=df8['MainBranch'].astype(str)
df8.plot(kind='barh', figsize=(12,8), color='lightcoral', alpha=.85)
plt.title('Bar Chart of Main Branch')
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.yticks(range(len(df8['MainBranch'])), labels)
for i, v in enumerate(df8['Num_of_Respondents']):
    plt.text(v + 0.1, i, str(v), color='black', va='center')

plt.show()
```
